```python
# first neural network with keras make predictions
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# load the dataset
TDC2 = loadtxt('NNTDtoC2.csv', delimiter=';')
TDC3 = loadtxt('NNTDtoC3.csv', delimiter=';')
testdata = loadtxt('NNTest1000run.csv', delimiter=';')
# split into input (X) and output (y) variables
X2 = TDC2[:,0:18]
y2 = TDC2[:,18]
X3 = TDC3[:,0:18]
y3 = TDC3[:,18]
X = testdata[:,0:18]
y = testdata[:,18]
# define the keras model
modelC2 = Sequential()
modelC2.add(Dense(4, input_shape=(18,), activation='relu'))
modelC2.add(Dense(1, activation='sigmoid'))
modelC3 = Sequential()
modelC3.add(Dense(4, input_shape=(18,), activation='relu'))
modelC3.add(Dense(1, activation='sigmoid'))
# compile the keras model
modelC2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
modelC3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
modelC2.fit(X2, y2, epochs=150, batch_size=10, verbose=0)
modelC3.fit(X3, y3, epochs=150, batch_size=10, verbose=0)
# make class predictions with the model
predictionsC2 = modelC2.predict(X)
predictionsC3 = (modelC3.predict(X) > 0.5).astype(int)
# summarize the first 5 cases
for i in range(50):
	#print("Recommendation to C2")
	print('%s => %f (expected %d)' % (X[i].tolist(), predictionsC2[i],y[i]))
	#print("Recommendation to C3")
	#print('%s => %d (expected %d)' % (X[i].tolist(), predictionsC3[i],y[i]))
```
